**percentile-Quantile**
- In the numpy package we have np.percentail() np.quatile()

- percentile: if you want to 25p 25

-

```python
num_cols=visa_df.select_dtypes(exclude='object').columns
dict1={}
for i in num_cols:
    wage_count=visa_df[i].count()
    wage_mean=round(visa_df[i].mean(),2)
    wage_median=round(visa_df[i].median(), 2)
    wage_max = round(visa_df[i].max(), 2)
    wage_min = round(visa_df[i].min(), 2)
    wage_std = round(visa_df[i].std(), 2)
    wage_25p = round(np.percentile(visa_df[i], 25), 2)
    wage_50p = round(np.percentile(visa_df[i], 50), 2)
    wage_75p = round(np.percentile(visa_df[i], 75), 2)
    list1 = [wage_count, wage_mean, wage_median, 
             wage_max, wage_min, wage_std, wage_25p,
             wage_50p, wage_75p]
    dict1[i]=list1
    
index_list = ['count', 'mean', 'median','max', 'min', 'std', '25%',
              '50%' ,'75%']
number_df=pd.DataFrame(dict1, index=index_list)
number_df.to_csv('number_visa.csv')
number_df
```

```python
p_wage = visa_df['prevailing_wage']
cont_25p = len(p_wage[p_wage<np.percentile(p_wage, 25)])
cont_50p = len(p_wage[p_wage<np.percentile(p_wage, 50)])
cont_75p = len(p_wage[p_wage<np.percentile(p_wage, 75)])

l1=['Count of 25%', 'Count_of 50%', 'Count_of_75%']
l2=[cont_25p, cont_50p, cont_75p]
d1=pd.DataFrame(zip(l1,l2), columns=['Til_per', 'Count'])
plt.bar('Til_per', 'Count', data=d1)
```
