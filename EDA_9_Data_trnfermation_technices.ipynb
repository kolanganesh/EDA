```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
exp_data = np.random.exponential(size=1000)
```

$Log-Transforms$

```python
log_data=np.log(exp_data)
log_data[:10]
```

```python
exp_data[:10]
```

```python
plt.hist(log_data, bins=40, label='log_data')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1).hist(exp_data, bins=40, label='Exponentioal')
plt.legend()
plt.subplot(1,2,2).hist(log_data, bins=40, label='Log_tranformation')
plt.legend()
plt.show()
```

$step-4$

**reciprocal tranformatnal**

- Reciprocal tranformation failed when data got zero value

```python
rec_data= 1/exp_data
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1).hist(exp_data, bins=40, label='Exponentioal')
plt.legend()
plt.subplot(1,2,2).hist(rec_data, bins=40, label='Reciprocal_tranformation')
plt.legend()
plt.show()
```

$step-5$

**Square root tranformation**

```python
squr_data = np.sqrt(exp_data)
squr_data[:10]
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1).hist(exp_data, bins=40, label='Exponentioal')
plt.legend()
plt.subplot(1,2,2).hist(squr_data, bins=40, label='Square root tranformation')
plt.legend()
plt.show()
```

```python
import seaborn as sns
sns.displot(squr_data)
plt.show()
```

$Step-6$

**power tranfermation**

- It is related to sklearn package

    - Package name: sklearn.preprocess
    

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer()
pwer_data= pt.fit_transform([exp_data])
pwer_data[:10]
plt.hist(pwer_data)
# plt.figure(figsize=(10, 5))
# plt.subplot(1,2,1).hist(exp_data, bins=40, label='Exponentioal')
# plt.legend()
# plt.subplot(1,2,2).hist(pwer_data, bins=40, label='Square root tranformation')
# plt.legend()
# plt.show()
```
